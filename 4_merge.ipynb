```python
import pandas as pd

grades = pd.read_csv('data/student/grades.csv')
student_information = pd.read_csv('data/student/student_information.csv')
unit = pd.read_csv('data/student/unit.csv')

display(grades)
display(student_information)
display(unit)
```

```python
combined_information = student_information.merge(how = 'left', right=grades, left_on='id', right_on='id')
display(combined_information)

all_combined = combined_information.merge(how = 'left', right=unit, left_on='unit', right_on='code')
display(all_combined)
```
